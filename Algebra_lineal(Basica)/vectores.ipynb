Ejercicio 1: Suma de vectores

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definimos los vectores
a = np.array([2, 3])
b = np.array([1, 2])
```

```python
# Calculamos la suma
c = a + b
```

```python
# Graficamos los vectores
plt.quiver(0, 0, a[0], a[1], color='b', label='Vector a')
plt.quiver(0, 0, b[0], b[1], color='g', label='Vector b')
plt.quiver(0, 0, c[0], c[1], color='r', label='Vector c (suma)')
plt.legend()
plt.grid(True)
plt.axis('equal') #Para que los ejes X e Y tengan la misma escala
plt.show()
```

Ejercicio 3: Producto escalar

```python
#Definimos los Vectores
a = np.array([2,3])
b = np.array([1,2])
```

```python
#Calculamos el Producto Escalar
producto = np.dot(a,b)
```

```python
#Graficamos los Vectores
plt.quiver(0, 0, a[0], a[1], color='b', label='Vector a')
plt.quiver(0, 0, b[0], b[1], color='g', label='Vector b')
plt.text(0.0, 0.0, f'Producto escalar: {producto:.2f}', ha='center')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

Ejercicios de resoluci√≥n de problemas
1.Un objeto se mueve desde el origen (0, 0) hasta el punto (3, 4) en un plano cartesiano. ¬øCu√°l es el vector que representa este movimiento?

```python
a = np.array([3,4])
plt.figure()
plt.quiver(3,4, color='purple', scale=7)
plt.title("El vector (3,4) indica el movimiento")
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid(True)
plt.show()
```

2.Dos fuerzas, F1 = (2, 3) y F2 = (1, 2), act√∫an sobre un objeto. ¬øCu√°l es el vector resultante de la suma de estas fuerzas?

```python
f1= np.array([2,3])
f2= np.array([1,2])
print(f1 + f2)
plt.figure()
plt.quiver(3,5, color='green', scale=8)
plt.title("El vector R de la suma de 2f = (3,5)")
plt.xlim(0,6)
plt.ylim(0,6)
plt.grid(True)
plt.show()
```

3.Un avi√≥n vuela desde el punto A (2, 3) hasta el punto B (5, 6). ¬øCu√°l es el vector que representa este movimiento?

```python
a = np.array([2,3])
b = np.array([5,6])
print(f'El vector que representa movimiento es C={a-b}')
plt.figure()
plt.quiver(0, 0, -3, -3, color='green', scale=6)
plt.xlim(-5, 0)
plt.ylim(-5, 0)
plt.grid(True)
plt.show()
```

Escribe un programa que calcule y grafique la suma de dos vectores a y b.

```python
vector_str1 = input("Ingrese las componentes del primer vector")
vector_str2 = input("Ingrese las componentes del segundo vector")
vector_list1 = vector_str1.split(',')
vector_list2 = vector_str2.split(',')
vector1 = [int(x) for x in vector_list1]
vector2 = [int(x) for x in vector_list2]
suma_de_vectores = np.add(vector1,vector2)
print(suma_de_vectores)
plt.figure()
plt.title(f'Suma de vectores:{suma_de_vectores}')
plt.quiver(suma_de_vectores[0],suma_de_vectores[1], color='red', scale=10)
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid(True)
plt.show()
```
